Day 3 Problem 1

Find all cases of the function mul(x, y) and return x * y.
Add the sum of all the multiplications.
Note - there are errors throughout text! Must discard all errors.

```python
with open('y2024d3_input.txt') as f:
    lines = f.readlines()

lines = ''.join(lines)
print(lines)
```

```python
import re
# Define the regex pattern to match 'mul(x,y)' where x and y are integers
pattern_mul = r"mul\((\-?\d+),(\-?\d+)\)"

matches = re.findall(pattern_mul, lines)
print(matches)
print(len(matches))
```

```python
sum = 0

for match in matches:
    sum += int(match[0]) * int(match[1])

print(sum)
```

Problem 2

Now must consider when mul() is enabled/disabled.

do() enables all following mul() functions.
don't() disables all following mul() functions.
Only the most recent do()/don't() applies.

Sum the outputs of only ENABLED mul() functions.
At the beginning, mul() is enabled.

```python
# find locations of all mul(), do(), and don't() in string
mul_idx = re.finditer(pattern_mul, lines)
#for match_mul in mul_idx:
#    print(match_mul.group())

pattern_do = r'\bdo\(\)'
do_idx = re.finditer(pattern_do, lines)

pattern_dont = r"\bdon't\(\)"
dont_idx = re.finditer(pattern_dont, lines)
#for match_dont in matches_dont:
#    print(match_dont.start(), match_dont.group())
```

```python
# combine all indices into one array (list)
# first column is index of where pattern found
# second column is multiple output (or 0 if not a mul row 
# third column is -1 for do, -2 for don't, 0 if mul row
import numpy as np


combined_list = []
for ind, data in enumerate(mul_idx):
    mul_output = int(int(matches[ind][0]) * int(matches[ind][1]))
    combined_list.append([data.start(), mul_output, 0])

for ind, data in enumerate(do_idx):
    combined_list.append([data.start(), 0, -1])

for ind, data in enumerate(dont_idx):
    combined_list.append([data.start(), 0, -2])

```

```python
# transfer to numpy array and sort by index
combined_array = np.array(combined_list, dtype=int)
combined_array = combined_array[combined_array[:,0].argsort()]
# add cols for operations
added_col = np.zeros((len(combined_array),1), dtype=int)
combined_array = np.hstack((combined_array, added_col))
```

```python
# if mul is enabled, copy value to third column
flag_enable = True

for i in range(len(combined_array)):
    if flag_enable:
        combined_array[i,3] = combined_array[i, 1]
    if combined_array[i, 2] == -1:
        flag_enable = True
    if combined_array[i, 2] == -2:
        flag_enable = False

print(np.sum(combined_array[:, 3]))
```
